# 基于pytorch的一个简单的线性回归模型

## 1.导入python包

```python
#下面两行代码是因为我的环境使用matplotlib内核会崩溃,正常环境可以去除
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
import numpy as np
import matplotlib.pyplot as plt
from torch import nn,optim
from torch.autograd import Variable
import torch
```

## 2.生成数据
下面使用随机生成的无意义数据,也可以采用其他有意义数据如房价,温度等

```python
x_data = np.random.rand(100)
noise = np.random.normal(0,0.01,x_data.shape)
y_data = x_data*0.1+0.2+noise
```

```python
plt.scatter(x_data,y_data)
plt.show()
```

```python
#使用reshape将数据转化为一维
x_data=x_data.reshape(-1,1) #-1表示自动匹配
y_data=y_data.reshape(-1,1)
```

```python
#将numpy数据转化为pytorch可以处理的tensor数据
x_data=torch.FloatTensor(x_data)
y_data=torch.FloatTensor(y_data)
inputs=Variable(x_data)
target=Variable(y_data)
```

## 3.构建神经网络模型

```python
class LinearRegression(nn.Module):
    #定义网络结构
    def __init__(self):
        #初始化父类nn.Module
        super().__init__()
        self.fc=nn.Linear(1,1)#输入特征一维,输出特征一维
    #定义前向传播
    def forward(self,x):
        return self.fc(x)
```

```python
#实例化一个线性回归模型
model =LinearRegression()
#定义损失函数为均方误差函数
mse_loss=nn.MSELoss()
#定义优化器为SGD,学习率为0.1
optimizer=optim.SGD(model.parameters(),lr=0.1)
```

```python
# 定义训练函数
def train(epoch):
    print("开始训练")
    for i in range(epoch):
        output = model(inputs)
        # 计算损失
        loss = mse_loss(output, target)
        # 梯度清零
        optimizer.zero_grad()
        # 计算梯度
        loss.backward()
        # 修正权值
        optimizer.step()
        if i % 200 == 0:  # 注意这里冒号是英文格式
            print(f'Epoch [{i+1}/{epoch}], Loss: {loss.item():.6f}')
```

## 4.进行模型训练
>理想情况,随着训练轮次的增加,loss值应该逐步减少并趋向稳定

```python
epoch=1001
train(epoch)
print("结束训练")
```

## 5.结果画图展示

```python
y_hat=model(inputs)
plt.scatter(x_data,y_data)
plt.plot(x_data,y_hat.data.numpy(),'r-',lw=3)#拟合结果的线展示为红色,宽度为3
plt.show()
```
